```python
import argparse
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from skimage.transform import rescale, resize, downscale_local_mean
import pickle
import os
```

```python
indir="./landscape"
outdir="./processed"
x = io.imread_collection(indir + "/*.jpg")
```

```python
shapes = []
for img_file in x.files:
    a = io.imread(img_file)
    if (a.shape[0] == 100 or a.shape[1] == 100) or (len(a.shape) !=3):
        print(img_file)
        new_file = "./removed/" + img_file.split('/')[-1]
        print(new_file)
        os.rename(img_file, new_file)
    else:
        shapes.append(a.shape)
```

```python
widths = [x[1] for x in shapes]
```

```python
heights = [x[0] for x in shapes]
```

```python
plt.figure(figsize=(20,20)) 
plt.plot(widths, heights, 'o')
```

```python
stats.describe(np.asarray(widths))
```

```python
stats.describe(np.asarray(heights))
```

```python
ex1 = io.imread(x.files[0])
ex1_resized = resize(ex1, (500, 500), anti_aliasing=True)
```

```python
def plot_resized(image, image_resized):
    fig, axes = plt.subplots(nrows=1, ncols=2,  figsize=(10,8))
    ax = axes.ravel()

    ax[0].imshow(image, cmap='gray')
    ax[0].set_title("Original image")

    ax[1].imshow(image_resized, cmap='gray')
    ax[1].set_title("Resized image (no aliasing)")
    plt.show()
```

```python
plot_resized(ex1, ex1_resized)
print(ex1.shape)
```

```python
ex2 = io.imread(x.files[1])
ex2_resized = resize(ex2, (500, 500), anti_aliasing=True)
```

```python
plot_resized(ex2, ex2_resized)
print(ex2.shape)
```

```python
ex3 = io.imread(x.files[2])
ex3_resized = resize(ex3, (500, 500), anti_aliasing=True)
plot_resized(ex3, ex3_resized)
print(ex3.shape)
```

```python
def scale_image(img, new_width, new_height):
    return resize(img, (new_width, new_height), anti_aliasing=True)
```

```python
resized_np = []
for img_file in x.files:
    a = io.imread(img_file)
    if (len(a.shape) == 3): # filter out gray-scale images
        a_resized = scale_image(a, 500, 500)
        resized_np.append(a_resized)
        out_file = outdir + "/" + img_file.split("/")[2]
        io.imsave(out_file, a_resized)
```

```python
resized_pickle_file = open("./resized_data", 'wb')
pickle.dump(resized_np, resized_pickle_file)
```
