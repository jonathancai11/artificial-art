```python
T = np.array(loss['idx'])
```

```python
fig = plt.figure()
fig.suptitle('Discriminator Loss', fontsize=15)

d_loss = np.array(loss['d_loss'])
x_d_new = np.linspace(T.min(),T.max(), 50) # number of points to make between T.min and T.max
d_loss_spl = make_interp_spline(T, d_loss, k=3)
d_loss_smooth = d_loss_spl(x_d_new)
plt.plot(x_d_new, d_loss_smooth)
```

```python
fig = plt.figure()
fig.suptitle('Generator Loss', fontsize=15)

g_loss = np.array(loss['g_loss'])
x_g_new = np.linspace(T.min(),T.max(), 300) 
g_loss_spl = make_interp_spline(T, g_loss, k=3)
g_loss_smooth = g_loss_spl(x_g_new)
plt.plot(x_g_new, g_loss_smooth)
```

```python
power = np.array(loss['g_loss'])
xnew = np.linspace(T.min(),T.max(), 300) 
spl = make_interp_spline(T, power, k=3)
power_smooth = spl(xnew)
plt.plot(xnew,power_smooth)
plt.show()
```
